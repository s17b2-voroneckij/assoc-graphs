```python
f = open('assoc_net.txt', encoding='utf-8').readlines()
```

```python
nums = []
for line in f:
    line = line.split(',')
    nums.append(float(line[-1]))
```

```python
np.average(nums)
```

```python
plt.figure(figsize=(15, 15))
plt.hist(nums, bins=1000)
plt.show()
```

```python
import numpy as np
```

```python
mu, sigma = 0.14, 0.1 # mean and standard deviation
s = np.minimum(np.random.exponential(0.15, 60000), np.array([100000000000000] * 60000))
print(np.average(s), s.shape)
```

```python
count, bins, ignored = plt.hist(s, 500, density=True, align='mid')
plt.show()
```

```python
mu, sigma = -2.0, 0.7 # mean and standard deviation
s = np.minimum(np.random.lognormal(mu, sigma, 60000), np.array([1] * 60000))
print(np.average(s), s.shape)
```

```python
count, bins, ignored = plt.hist(s, 500, density=True, align='mid')
plt.show()
```

```python
s = np.random.gamma(0.25, 0.7, 60000)
print(np.average(s))
count, bins, ignored = plt.hist(s, 200, density=True, align='mid')
plt.show()
```

```python
s = np.random.beta(0.025, 0.25, 60000)
print(np.average(s))
count, bins, ignored = plt.hist(s, 200, density=True, align='mid')
plt.show()
```

```python
from random import randint
my_generator = lambda: nums[randint(0, len(nums) - 1)]
s = [my_generator() for i in range(600000)]
print(np.average(s))
plt.figure(figsize=(15, 15))
count, bins, ignored = plt.hist(s, 500, density=True, align='mid')
plt.show()
```
